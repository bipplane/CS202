# Implementation of Insertion Sort and Merge Sort

```python
def insertion_sort(A):
    # A is a list of numbers
    # write your own code to implement insertion sort
    for i in range(1, len(A)):
        value = A[i]
        j = i - 1
        while j >= 0 and A[j] > value:
            A[j + 1] = A[j]
            j -= 1
        A[j + 1] = value
        print(A)
    return A
        

def merge_sort(A):
    # A is a list of numbers
    # write your own code to implement merge sort with recursion
    def merge(L, R, A):
        i = j = 0
        for k in range(len(A)):
            if L[i] <= R[j]:
                A[k] = L[i]
                i += 1
                if i == len(L):
                    A[k + 1:] = R[j:]
                    break
            else:
                A[k] = R[j]
                j += 1
                if j == len(R):
                    A[k + 1:] = L[i:]
                    break


    if len(A) > 1:
        mid = len(A) // 2
        Lhalf = A[:mid]
        Rhalf = A[mid:]

        merge_sort(Lhalf)
        merge_sort(Rhalf)
        merge(Lhalf, Rhalf, A)

    return A
```

```python
import time
import random
list1 = [random.randint(0, 100) for _ in range(100)]
list2 = [random.randint(0, 10000) for _ in range(10000)]
list3 = [random.randint(0, 1000000) for _ in range(100000)]

start_time = time.time()
merge_sort(list1.copy())
print("--- %s seconds spent by merge sort for list1 ---" % (time.time() - start_time))
start_time = time.time()
insertion_sort(list1.copy())
print("--- %s seconds spent by insertion sort on list1 ---" % (time.time() - start_time))


start_time = time.time()
merge_sort(list2.copy())
print("--- %s seconds spent by merge sort for list2 ---" % (time.time() - start_time))
start_time = time.time()
insertion_sort(list2.copy())
print("--- %s seconds spent by insertion sort on list2 ---" % (time.time() - start_time))




start_time = time.time()
merge_sort(list3.copy())
print("--- %s seconds spent by merge sort for list3 ---" % (time.time() - start_time))
start_time = time.time()
insertion_sort(list3.copy())
print("--- %s seconds spent by insertion sort on list3 ---" % (time.time() - start_time))
```
