```python
import itertools

def conflict(i1, j1, i2, j2):
    return (i1-i2 == j1-j2) or (i1 - i2 == - (j1-j2))
# this is the check function
# should output if two queens threaten each other

def n_queen(n):
    num_ways = 0

    # TODO: how to list all possible solutions?
    perms = (itertools.permutations(range(n)))
    for i in perms:
        valid = True
        for j in range(len(i)):
            for k in range(j+1, len(i)):
                if conflict(j, i[j], k, i[k]):
                    valid = False
                    break
            if not valid:
                break
        if valid:
            num_ways += 1


    return num_ways

if __name__ == '__main__':

    n = 8

    while True:
        try:
            if n > 0:
                break
        except ValueError:
            print('try harder!')
    print('the number of distinct solutions is %d.' % n_queen(n))
```
