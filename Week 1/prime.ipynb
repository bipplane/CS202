```python
import math, time

def naive(upper_bound):
    prime = [2]
    # i: for all odd numbers less than upper_bound
        # j: for all odd numbers less than i
    for i in range(3, upper_bound, 2):
        for j in range(3, i, 2):
            if i % j == 0:
                break # why break?
        else:
            prime.append(i)
    return prime

def improved(upper_bound):
    prime = [2]
    for i in range(3, upper_bound, 2):
        found = False
        # TODO: How to implement the improved version? What is the largest number to check?
        for j in range(3, i, 2):
          if j > math.sqrt(i): break
          if i % j == 0: found = True
        if not found:
            prime.append(i)
    return prime
```

```python
upper_bound = 400
naive(upper_bound)
```
