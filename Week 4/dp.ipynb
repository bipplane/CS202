# Exercise 1: Coin Change with Unlimited Supply (Iterative Implementation)

```python
# implement an algorithm to compute the minimum number of coins to make up a given change, assuming the supply of coins for each denomination are unlimited

# with denomination [1, 3, 4] cents, what is the minimum number of coins to make up 30 cents?
# an example with only 3 denominations: 1-cent, 3-cents and 4-cents
# suppose there is unlimited supply for each denomination
# what is the minimum number of coins to make up 30 cents
# is it 8? 7 4-cents + 2 1-cent

denom = [1, 3, 4]
m = len(denom)
n = 30

min_coin = [float('inf')] * (n + 1)
min_coin[0] = 0

for i in range(1, n + 1):
    for j in range(m):
        if denom[j] <= i:
            min_coin[i] = min(min_coin[i], min_coin[i - denom[j]] + 1)


for e in enumerate(min_coin):
    print(e)
```

# Exercise 2: Coin Change with Unlimited Supply (Recursive Implementation)

```python
denom = [1, 3, 4]
m = len(denom)
n = 30

min_coin = {} # dict
def coin_change_unlimited_recur_memo(i):
    if i == 0:
        min_coin[0] = 0
        return 0
    if i in min_coin:
        return min_coin[i]
    else:
        cur_min = float('inf')
        for d in denom:
            if i - d >= 0:
                cur_min = min(cur_min, coin_change_unlimited_recur_memo(i - d) + 1)

        min_coin[i] = cur_min
        return min_coin[i]

n = 100
coin_change_unlimited_recur_memo(n)

print(min_coin)
```
